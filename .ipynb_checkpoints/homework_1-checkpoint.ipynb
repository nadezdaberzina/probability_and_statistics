# Теория вероятностей и математическая статистика

## Урок 1. Случайные события. Условная вероятность. Формула Байеса. Независимые испытания

```python
from math import factorial
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

### Задание 1

#### Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

##### а) Вероятность того, что все карты – крести

```python
C_clovers = combinations(13, 4)
C_total = combinations(52, 4)
P = C_clovers / C_total
```

```python
print(f'Количество возможных комбинаций с картами крести: {C_clovers}')
print(f'Общее число исходов: {C_total}')
print(f'Вероятность того, что все карты - крести: {P} или около 0.26%')
```

##### б) Вероятность, что среди 4-х карт окажется хотя бы один туз

```python
C1 = combinations(4, 1) * combinations(48, 3)
C2 = combinations(4, 2) * combinations(48, 2)
C3 = combinations(4, 3) * combinations(48, 1)
C4 = 1
C = C1 + C2 + C3 + C4
```

```python
print(f'Количество возможных комбинаций с 1-м тузом и 3-мя картами: {C1}')
print(f'Количество возможных комбинаций с 2-мя тузами и 2-мя картами: {C2}')
print(f'Количество возможных комбинаций с 3-мя тузами и 1-ой картой: {C3}')
print(f'Количество возможных комбинаций с 4-мя тузами: {C4}')
print(f'Количество возможных комбинаций из 4-х карт с хотя бы одним тузом: {C}')
print(f'Общее число исходов: {C_total}')
print(f'Вероятность, что среди 4-х карт окажется хотя бы один туз: {C / C_total} или около 28%')
```

### Задание 2

#### На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

```python
C = combinations(10, 3)
```

```python
print(f'Количество возможных комбинаций из 3-х цифр: {C}')
print(f'Вероятность того, что человек, не знающий код, откроет дверь с первой попытки: {1 / C} или около 0.83%')
```

### Задание 3

####  В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

```python
P1 = 9 / 15
P2 = 8 / 14
P3 = 7 / 13
P = P1 * P2 * P3
```

```python
print(f'Вероятность того, что 1-ая деталь будет окрашена: {P1}')
print(f'Вероятность того, что 2-ая деталь будет окрашена: {P2}')
print(f'Вероятность того, что 3-я деталь будет окрашена: {P3}')
print(f'Вероятность того, что все извлеченные детали окрашены: {P} или около 18%')
```

### Задание 4

#### В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

```python
P1 = 2 / 100
P2 = 1 / 99
P = P1 * P2
```

```python
print(f'Вероятность того, что 1-ый билет окажется выигрышным: {P1}')
print(f'Вероятность того, что 2-ой билет окажется выигрышным: {P2}')
print(f'Вероятность того, что два билета окажутся выигрышными: {P} или около 0.02%')
```
